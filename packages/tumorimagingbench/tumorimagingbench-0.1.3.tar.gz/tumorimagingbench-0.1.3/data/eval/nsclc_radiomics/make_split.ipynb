```python
import pandas as pd
```

```python
df = pd.read_csv("annotations.csv")
```

```python
df
```

```python
# Split the data into training and validation sets
from sklearn.model_selection import train_test_split

train_df, val_test_df = train_test_split(df, test_size=0.5)
val_df, test_df = train_test_split(val_test_df, test_size=0.8)

train_df.to_csv("train_annotations.csv", index=False)
val_df.to_csv("val_annotations.csv", index=False)
test_df.to_csv("test_annotations.csv", index=False)
```
